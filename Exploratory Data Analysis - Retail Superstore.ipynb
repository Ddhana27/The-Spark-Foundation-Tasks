# Perform Data Visualizations

# Pair Plot to have a glance visualization at the entire dataset

```python
sns.pairplot(data = df)    #It will plot each field with respect another.
```

# Line Plot for Sales w.r.t Profit in view of 'Ship Mode'

```python
sns.lineplotLine Plot for Sales w.r.t Profit in view of 'Ship Mode'(x = df["Sales"] , y = df["Profit"], hue = df["Category"], color = 'g')
```

# Analysis-1 :
Through the Line Plot, we've found that, with increase in 'Sales' the 'Profit' is not clearly increasing. It is randomly moving up and down.
Bar Plot for 'Discount' w.r.t 'Sales' in view of 'Profit'

```python
sns.scatterplot(x = df["Discount"], y = df["Profit"], hue = df["Category"])
```

# Analysis-2
After this analysis, we came to know that, the increase in 'Discount' means the decrease in 'Profit'. The problem may be because the more 'Discount' is given, which directly decreased the 'Profit'.
Count Plot to know the ratio of Sub-Categories w.r.t each other¶

```python
plot = sns.countplot(df["Sub-Category"])
plot.set_xticklabels(plot.get_xticklabels() , rotation = 90)
```

# Analysis-3
This Analysis shows that 'Binders' have highest Sales, then 'Papers' and then 'Furnishings'.
Bar Chart for Ship_Mode categories

```python
plot = sns.countplot(df["Ship Mode"])
plot.set_xticklabels(plot.get_xticklabels() , rotation = 40)
```

# Analysis-4
In the Ship Mode, the 'Standard Class' is preferred more than other modes.
Bar Chart for Segment categories

```python
plot = sns.countplot(df["Segment"])
plot.set_xticklabels(plot.get_xticklabels() , rotation = 10)
```

# Analysis-5
The Sales related to 'Consumer' has more counts than 'Corporate' and 'Home Office'.
Bar Chart for 'Category'

```python
plot = sns.countplot(df["Category"])
plot.set_xticklabels(plot.get_xticklabels() , rotation = 10)
```

# Analysis-6
The 'Office Supplies' are more than 'Furniture' and 'Technology'. There is need to take steps to increase the Sales of 'Furniture' and 'Technology'.
Bar Chart for Region based Supplies

```python
plot = sns.countplot(df["Region"])
plot.set_xticklabels(plot.get_xticklabels() , rotation = 20)
```

# Analysis-7
There are more orders from the 'West' and 'East ' of United States than 'South' and 'Central' areas. We should engage more on 'South' and 'Central' areas of The United States.¶

```python
sns.barplot(x = df["Category"] , y = df["Discount"])
```

# Analysis-8
If we increase the 'Discount' from 0.5, the 'Profit' decrease eventually.
Conclusion
1) Discount should be given thoughtfully, unnecessary discount witnesses low profit.
2) Discount on Furniute Supplies should be kept low, as we are facing more loss here.
3) We should concentrate on Technology Sales because we have low Sales in it.
4) The Sales at the South and Central part of United States is lower than other Regions. More work should be done there to increase the number of Sales.
5) People prefer 'Standard Class' ship mode more than other ship modes.
